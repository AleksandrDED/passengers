```python
import pandas as pnd
import matplotlib.pyplot as plot
import numpy as npy
from statsmodels.tsa.holtwinters import ExponentialSmoothing as HWES
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error as mse
```

```python
# Загрузка данных из CSV файла
dataset = pnd.read_csv('AirPassengers.csv', index_col='Month', parse_dates=True)
```

```python

# Построение графика исходных данных с использованием точек
plot.figure(figsize=(10, 6))
plot.plot(dataset, marker='o', linestyle='-', color='b')
plot.xlabel('Дата')
plot.ylabel('Кол-во пассажиров')
plot.title('Ежемесячные данные о транспортировке')
plot.grid(True)
plot.show()
```

```python
# Расчёт первого и третьего квартилей и межквартильного размаха
first_quartile = dataset['#Passengers'].quantile(0.25)
third_quartile = dataset['#Passengers'].quantile(0.75)
interquartile_range = third_quartile - first_quartile
```

```python
# Определение нижней и верхней границ для удаления выбросов
low_bound = first_quartile - 1.5 * interquartile_range
up_bound = third_quartile + 1.5 * interquartile_range
```

```python
# Фильтрация данных для удаления выбросов
dataset = dataset[(dataset['#Passengers'] >= low_bound) & (dataset['#Passengers'] <= up_bound)]
```

```python
# Построение графика данных после удаления выбросов с использованием штриховой линии
plot.figure(figsize=(10, 6))
plot.plot(dataset, linestyle='--', color='r')
plot.xlabel('Дата')
plot.ylabel('Количество пассажиров')
plot.title('Ежемесячные данные о транспортировке после удаления аномалий')
plot.grid(True)
plot.show()
```

```python
# Разделение данных на обучающую и тестовую выборки
training_data, testing_data = tts(dataset, test_size=0.2, shuffle=False)
```

```python
# Создание модели экспоненциального сглаживания
hwes_model = HWES(training_data['#Passengers'], seasonal='add', seasonal_periods=12)
```

```python
# Обучение модели на обучающей выборке
fit_hwes_model = hwes_model.fit()
```

```python
# Прогнозирование на тестовой выборке
forecast_test = fit_hwes_model.forecast(len(testing_data))
```

```python
# Прогнозирование на будущее (12 месяцев вперёд)
forecast_future = fit_hwes_model.forecast(12)
```

```python
# Построение графика обучающей, тестовой выборок и прогнозов с разными цветами и маркерами
plot.figure(figsize=(12, 8))
plot.plot(training_data.index, training_data['#Passengers'], label='Обучающая выборка', marker='o', linestyle='-', color='g')
plot.plot(testing_data.index, testing_data['#Passengers'], label='Тестовая выборка', marker='s', linestyle='-', color='b')
plot.plot(testing_data.index, forecast_test, label='Прогноз на тестовой выборке', marker='^', linestyle='-', color='orange')
plot.plot(pnd.date_range(start=testing_data.index[-1], periods=13, freq='MS')[1:], forecast_future, label='Прогноз на следующий год', marker='x', linestyle='-', color='purple')
plot.xlabel('Дата')
plot.ylabel('Количество пассажиров')
plot.title('Прогноз пассажиропотока')
plot.legend()
plot.grid(True)
plot.show()
```

```python
# Расчёт среднеквадратичной ошибки на тестовой выборке
mean_sq_error = mse(testing_data['#Passengers'], forecast_test)
print(f'Среднеквадратичная ошибка на тестовой выборке: {mean_sq_error}')
```

```python
# Построение графика фактических и прогнозных данных на тестовой выборке с использованием точек и штриховой линии
plot.figure(figsize=(10, 6))
plot.plot(testing_data.index, testing_data['#Passengers'], label='Фактические данные', marker='o', linestyle='-', color='r')
plot.plot(testing_data.index, forecast_test, label='Прогноз', marker='d', linestyle='--', color='blue')
plot.xlabel('Дата')
plot.ylabel('Количество пассажиров')
plot.title('Фактические и Прогнозные данные о транспортировке')
plot.legend()
plot.grid(True)
plot.show()
```
